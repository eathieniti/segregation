# Schelling Segregation Model

## Background

The Schelling (1971) segregation model is a classic of agent-based modeling, demonstrating how agents following simple rules lead to the emergence of qualitatively different macro-level outcomes. Agents are randomly placed on a grid. There are two types of agents, one constituting the majority and the other the minority. All agents want a certain number (generally, 3) of their 8 surrounding neighbors to be of the same type in order for them to be happy. Unhappy agents will move to a random available grid space. While individual agents do not have a preference for a segregated outcome (e.g. they would be happy with 3 similar neighbors and 5 different ones), the aggregate outcome is nevertheless heavily segregated.

```python


def get_filename_pattern(factor,num_steps, minority_pc, M0, M1, temp,height,width,
        move,symmetric_positions, residential_steps,alpha, density,schelling,
        school_moves_per_step, residential_moves_per_step, bounded, radius, cap_max, T,fs, variable_f,sample,sigma, num_neighbourhoods, schools_per_neighbourhood, displacement,b,f0,f1,version):

    fs_print = fs
    if fs_print=="eq":
        fs_print= 0

    #version="V3_fx2_"
    #version="V4_"
    if factor =='f0':
        filename_pattern=version + "%s_m=%.2f_M0=%.2f_M1=%.2f_temp_%.2f_h_%d_st_%d_move_%s_sym_%s_res_%d_a_%.2f_den_%.2f_schell_%s_s_mps_%d_r_mps_%d_bounded_%s_r_%d_cp_%.2f_T_%.2f_fs_%.2f_v%s_s%d_sig%.2f_n%d_sn%d_d%d_b%.2f"%(
            factor,minority_pc, M0, M1, temp,height, num_steps,
        move,symmetric_positions, residential_steps,alpha, density,schelling,
        school_moves_per_step, residential_moves_per_step, bounded, radius, cap_max, T,fs_print, str(variable_f)[0],sample,sigma, num_neighbourhoods, schools_per_neighbourhood, displacement,b)
   
    return(filename_pattern)


def preprocessing_df(all_models_df):
    all_models_df=all_models_df.reset_index()
    all_models_df=all_models_df.sort_values(by=['iter','iter1',factor,'Step'])
    
    
    all_models_df['school_step'] = all_models_df['Step'] -all_models_df['res']
    all_models_df_filtered = all_models_df[all_models_df['res']>=0]

    all_models_df_filtered.res_seg_index = pd.to_numeric(all_models_df_filtered.res_seg_index, errors='coerce').fillna(0)
    return(all_models_df_filtered)
    
    
def get_values(all_models_df):
    all_models_df_filtered1 = all_models_df[all_models_df.res>0]

    indexes = []
    for ind2, model_out_dummy in all_models_df_filtered1.groupby(['res', vary]):

        for ind, model_out in model_out_dummy.groupby(['iter','iter1', factor]):
            indexes.append(ind[2])

        print(ind2,Counter(indexes))

    print(all_models_df[vary].unique())
    
    
def load_dfs(params_set):
    dfs1=[]
    filename_pattern = get_filename_pattern(factor=factor, num_steps=num_steps,**params_set,version=version )
    filename_pattern_to_find= filename_pattern + '*'
    i=0

    for name in glob.glob("dataframes/*odels*" + filename_pattern_to_find):
        #try:
            print("file found", name)
            new_df = pd.read_pickle(name)
            new_df["res"] = np.repeat(params_set['residential_steps'], len(new_df))
            new_df["iter1"] = np.repeat(i, len(new_df))
            new_df[vary] = np.repeat(params_set[vary], len(new_df))
            dfs1.append(new_df)

            i+=1
    print("found %d iterations"%i)
    return(dfs1)
```
